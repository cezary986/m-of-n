## Comparison with Decision Tree

```python
Tables.configure('../survival/decision_tree')
decision_tree_metrics = pd.concat(Tables.query('*', 'metrics', as_pandas=True))
columns: list = decision_tree_metrics.columns.tolist()
# columns.remove('dataset')
# columns.insert(0, 'dataset')
decision_tree_metrics = decision_tree_metrics[columns]

decision_tree_metrics = decision_tree_metrics.set_index('dataset')
rules_metrics = metrics.set_index('dataset')


pd.DataFrame({
    'rules IBS (train)': rules_metrics['integrated_brier_score (train)'],
    'model_type': rules_metrics['model_type'],
    'DC IBS (train)': decision_tree_metrics['integrated_brier_score (train)'],
    'rules IBS (test)': rules_metrics['integrated_brier_score (test)'],
    'DC IBS (test)': decision_tree_metrics['integrated_brier_score (test)'],
}, index=rules_metrics.index)
```

```python
agg_metrics = pd.DataFrame({
    'rules IBS (train)': rules_metrics['integrated_brier_score (train)'],
    'model_type': rules_metrics['model_type'],
    'rules IBS (test)': rules_metrics['integrated_brier_score (test)'],
}, index=rules_metrics.index).groupby('model_type').agg(string_supporting_mean)

agg_metrics['DC IBS (train)'] = decision_tree_metrics['integrated_brier_score (train)'].mean()
agg_metrics['DC IBS (train) (std)'] = decision_tree_metrics['integrated_brier_score (train)'].std()
agg_metrics['DC IBS (test)'] = decision_tree_metrics['integrated_brier_score (test)'].mean()
agg_metrics['DC IBS (test) (std)'] = decision_tree_metrics['integrated_brier_score (test)'].std()

agg_metrics
```
